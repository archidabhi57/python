### 01) WAP to demonstrate the use of Pie chart.

```python
import matplotlib.pyplot as plt
%matplotlib notebook
l=['Noodles','Panipuri','Pavbhaji','Manchurian']
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0, 0.1, 0) 

plt.pie(sizes, explode=explode,colors=colors,labels=l,
        shadow=True,
 autopct='%1.1f%%',startangle=140)

plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
import matplotlib.pyplot as plt
%matplotlib notebook

x=[1,2,3,4,5]
y=[5.9,3.2,1.2,9.7,2.2]
l=['1st','2nd','3rd','4th','5th']

c=['b','g','r','c','m']
w=[0.5,0.3,0.2,0.7,0.9]
plt.bar(x,y,color=c,label=l,width=w)
plt.title('Trade view')


plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np

x=[1,2,3,4,5]
y=[5.9,3.2,1.2,9.7,2.2]
l=['1st','2nd','3rd','4th','5th']
ind = np.arange(len(y))

fig,ax=plt.subplots()

c=['b','g','r','c','m']
for i,v in enumerate(y):
    ax.text(v+0,i+.15,str(v),
           color='blue',fontweight='bold')
plt.barh(ind,y,color=c,label=l)
plt.title('Trade view')


plt.show()
```
